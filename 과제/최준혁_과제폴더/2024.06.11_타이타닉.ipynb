```python
# 12. 'Age' 필드의 결측치는 값이 중간값으로 할당하기
age_med = df['Age'].median()
```

```python
df['Age'].fillna(age_med, inplace=True)
df.isna().sum()
```

```python
# 13. 'Name'필드에서 신분을 나타내는 단어를 뽑아서 'Title' 필드에 할당하기
df['Name'].head()
# Mrs., Mr., Miss. 등등..
# 공통점으론 영문 + . 이 있다.
```

```python
res =df['Name'].str.extract('([A-Za-z]+)\.', expand=False)
res
```

```python
df['Title'] = res
df.head()
```

```python
# 14. 'Title' 필드의 카테고리를 6개의 항목으로 변경하기
df['Title']
```

```python
df['Title'].value_counts()
```

```python
df['Title'].unique()
```

```python
def categorize(title):
    if title in ['Mr']:
        return 'Mr'
    elif title in ['Miss', 'Mlle', 'Ms']:
        return 'Miss'
    elif title in ['Mrs', 'Mme']:
        return 'Mrs'
    elif title in ['Master']:
        return 'Master'
    elif title in ['Dr', 'Rev']:
        return 'Dr/Rev'
    else:
        return 'Other'
```

```python
df['Title'] = df['Title'].apply(categorize)
df['Title'].value_counts()
```

```python
# prompt: # 16. 나이('Age') 필드를 그룹핑하여 'AgeGroup'필드 생성하여 할당하기

def age_group(age):
  if age < 18:
    return 'Child'
  elif age >= 18 and age <= 35:
    return 'Young Adult'
  elif age > 35 and age <= 50:
    return 'Adult'
  else:
    return 'Senior'

df['AgeGroup'] = df['Age'].apply(age_group)
df.head()
```
